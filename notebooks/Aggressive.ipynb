```python
import pandas as pd
import numpy as np
import spacy

import gensim
from gensim.models import Word2Vec 
import gensim.downloader

```

```python
#import lexicon with and assign the right column names
df = pd.read_csv("/content/data_sets/NRC-VAD-Lexicon.txt", sep='\t', header = None)
col = ['valence', 'arousal', 'dominance']
df = df.rename(columns={1: col[0] , 2: col[1] , 3: col[2]})
```

```python
#evaluate the aggressivity score of each word in the lexicon as word_arousal / word_valence
aggressivity = []
for arousal, valence in zip ( list(df['arousal']) , list(df['valence']) ):
  aggressivity.append(arousal / (valence + 0.01)) 

df['aggressivity'] = aggressivity
```

```python
#download with gensim "glove-twitter-100" for vectorial words representations
glove_vectors = gensim.downloader.load('glove-twitter-100')
```

```python
glove_vectors.similarity('hi', 'hello')
```

```python
#define a function that given a word w, the reference lexicon dataset and the glove model
#finds the best matching w' in the lexicon having the highest cosine similarity score when compared to w


def find_best_match(word, df, model):
  best_match = df[0][0]
  current_cosine = 0.0
  for w in df[0]:
    try:
      if model.similarity(w, word) > current_cosine:
        current_cosine = model.similarity(w, word)
        best_match = w
    except:
      continue

  return best_match, current_cosine

 
```

```python
find_best_match('facetious', df, glove_vectors)
```

```python
#define a function that given a word w, our lexicon and the chosen glove model, outputs a tuple
# (aggressivity, cosine) where "aggressivity" is the aggressivity value assigned to w', being the best matching
#word in the lexicon related to w according to find_best_match(), and "cosine" i the cosine similairty value between w and w'

def aggressiveness(word, df, model):
  best_match, cosine = find_best_match(word, df, model )

  aggressivity = float(df[df[0] == best_match]['aggressivity'])
  if cosine == 0:
    aggressivity = 0
  return (aggressivity, cosine)
```

```python
#import labeled tweet data-set, already normalized
df_tweets = pd.read_csv("/content/data_sets/all_annotated_tweets_update.csv", sep=',', header=0)
df_tweets

```

```python
#define a dictionary having "occupation" labels as keys and as values the respective sub-dataframe of df_tweet
#including all and only recording matching their occupation label with the key taken into account

dataframes = {}

for occupation in set(df_tweets['occupation']):
  dataframes[occupation] = df_tweets[df_tweets['occupation'] == occupation]
```

```python
dataframes['arts']
```

```python
to_exclude = ['@USER','HTTPURL','']
nlp = spacy.load('en_core_web_sm')

#for each sub-dataframe, perform a proper normalization, namely:
#for each tweet in the sub-dataframe, exclude '@USER','HTTPURL',''
#consider just the lemmas
#exclude stop words and named entities and punctuation; exclude lemmas having a length of 1


for key, df_ in dataframes.items():
  tokenized = []

  for i in df_['text']:
    text_data = i
    doc = nlp(text_data)
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.ent_type_ and not token.is_punct]
    tokens = [tok.lower() for tok in tokens if tok not in to_exclude]
    tokens = [tok for tok in tokens if len(tok) > 1]
    tokenized.append((i,tokens))

    #substitute the dataframe with the list of tokenized tweets

    dataframes[key] = tokenized
```

```python
dataframes['arts']
```

```python
#given a tokenized string, evaluate its aggressive value and reliability value respectively as
#the mean aggressive value and the mean reliability value

def extract_aggressiveness(tokenized_string):
  aggr_values = [aggressiveness(x, df, glove_vectors)[0] for x in tokenized_string]
  cosine_values = [aggressiveness(x, df, glove_vectors)[1] for x in tokenized_string]
  aggr_values = np.mean(np.array(aggr_values))
  cosine_values = np.mean(np.array(cosine_values))
  return aggr_values, cosine_values
```

```python
#for each sub-dataframe, evaluate the aggressiveness score and the respective reliability of each tweet
#save in a list "levels" only values having a reliability > 0.7 then use the list as value of the results dictionary
#in relation to the respective key

#NOTE: computation for this cell block takes a lot of time

results = {}

for key, value in dataframes.items():
  levels = []
  for tup in value:
    tokenized_string = tup[1]
    aggr_value , reliability = extract_aggressiveness(tokenized_string)
    if reliability < 0.7:
      continue
    levels.append((tup, aggr_value, reliability))
    
    
  results[key] = levels
```

```python
#list top 10 tweets in terms of aggressiveness for each category

for occupation in list(dataframes.keys()):  
  results[occupation].sort(key=lambda tup: tup[1], reverse=True) 
  print(occupation)
  print()
  for x in results[occupation][0:10]:
    print(x)
  print('\n\n')
```

```python
#for each occupation, display the mean aggressivity score considering the whole sub-dataframes

for occupation in list(dataframes.keys()):  
  mean_score = 0
  for x in (results[occupation]):
    if x[1] > -1: #needed to handle nan values which give to the expression a false truth value
      mean_score += x[1] 
    else:
      continue     
  print(occupation, mean_score / len(results[occupation]))
  
  
```
